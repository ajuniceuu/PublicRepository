```python
import json
import time
import urllib.request
import pandas

latitude_list=[]
longitude_list=[]
timestamps_list=[]

for r in range(125):
    url="http://api.open-notify.org/iss-now.json"
    response= urllib.request.urlopen(url)
    output= json.loads(response.read())
    
    location=output["iss_position"]
    lat=location["latitude"]
    lat=float(lat)
    
    long=location["longitude"]
    long=float(long)
    
    timestamp=output["timestamp"]
    
    latitude_list.append(lat)
    longitude_list.append(long)
    timestamps_list.append(timestamp)
    
    time.sleep(10)
    
d= {"lat":latitude_list,"long":longitude_list,"time":timestamps_list}
df= pandas.DataFrame(data=d)
df
```

```python
import matplotlib.pyplot as plot

fig = plot.figure(figsize=(40,15))
ax=fig.add_subplot()
ax.scatter(latitude_list,longitude_list,color="blue")
plot.xlabel("Latitude")
plot.ylabel("Longitude")
plot.title("International Space Station Position")
ax.set_xlim([-10,75])
ax.set_ylim([-175,175])
```
